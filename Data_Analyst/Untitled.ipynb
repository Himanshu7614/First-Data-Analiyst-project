```python
df['month_year']= df['order_date'].apply(lambda x: x.strftime('%y-%m'))
```

```python
df['month_year']
```

```python
df.groupby('month_year').sum()
```

```python
#Grouping Month year From the dataset
df_monthly_sale = df.groupby('month_year').sum()['sales'].reset_index()
```

```python
df_monthly_sale
```

(HOW CAN WE CREAT GRAPHS USING MATPLOTLIP, TITLE)

```python
plt.figure(figsize=(16,7))
plt.plot(df_monthly_sale['month_year'],df_monthly_sale['sales'])
plt.xticks(rotation='vertical',size=9)
plt.show()
```

<h2> How sale product wise

```python
df_product_sale = pd.DataFrame(df.groupby('product_name').sum()['sales'])
```

```python
best_selling_prods = df_product_sale.sort_values('sales',ascending=False)
```

```python
most_selling_product = pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```
